```python
import os, aifc, scipy
import numpy as np
from scipy import fftpack
from scipy import *

wave_path= "/Users/sayantanm/Downloads/sound_files"
labels = [f for f in os.listdir(wave_path)]

for i in labels:
    print '/Users/sayantanm/Downloads/sound_files/'+ i
    
    f=aifc.open('/Users/sayantanm/Downloads/sound_files/'+ i)
    
    
    params= list(f.getparams())
    
    strsig = f.readframes(f.getnframes())
    y = np.fromstring(strsig, dtype=np.int32)
    
    time = arange(size(y)) / params[2]
    freqs = scipy.fftpack.fftfreq(y.size, d=1/33.34)
    
    note= {'C0': 16,'C0/D0': 17, 'D0': 18,'D0/E0': 19,'E0': 20}
    note
    notemax= int(max(freqs))
    for i in range(len(note)):
        if notemax==note.values()[i]:
            x= note.keys()[i]
            print "Peak is at is : ",notemax,"and the note is : ",x
    fig = plt.figure(figsize=(16, 8))
    ax1 = fig.add_subplot(1,2,1)
    ax1.plot(time, freqs, color="blue", linestyle="-")
    ax1.set_xlabel("Time [s]")
    ax1.set_ylabel("Frequency")
    ax2 = fig.add_subplot(1,2,2)
    ax2.plot(time, y, color="red", linestyle="-")
    ax2.set_xlabel("Time [s]")
    ax2.set_ylabel("Amplitude")
```
